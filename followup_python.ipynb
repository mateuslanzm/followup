## 1. Variáveis e Tipos de Dados

```python
# Tipos básicos em Python
nome = "Python"
versao = 3.11
ativo = True
quantidade = 42

print(f"Linguagem: {nome}")
print(f"Versão: {versao}")
print(f"Ativo: {ativo}")
print(f"Quantidade: {quantidade}")
print(f"Tipos: {type(nome)}, {type(versao)}, {type(ativo)}, {type(quantidade)}")
```

## 2. Estruturas de Dados

```python
# Lista
linguagens = ["Python", "JavaScript", "Java", "C++", "Go"]
print("Lista:", linguagens)
print("Primeira:", linguagens[0])
print("Última:", linguagens[-1])

# Tupla
coordenadas = (10.5, -23.8)
print("\nTupla:", coordenadas)

# Dicionário
pessoa = {
    "nome": "Mateus",
    "profissao": "Desenvolvedor",
    "linguagem_favorita": "Python"
}
print("\nDicionário:", pessoa)

# Conjunto
numeros_unicos = {1, 2, 3, 2, 1, 4}
print("\nConjunto (sem duplicatas):", numeros_unicos)
```

## 3. Controle de Fluxo

```python
# Condicional
nota = 8.5

if nota >= 9.0:
    conceito = "A"
elif nota >= 7.0:
    conceito = "B"
elif nota >= 5.0:
    conceito = "C"
else:
    conceito = "D"

print(f"Nota: {nota} → Conceito: {conceito}")

# Loop for
print("\nLinguagens populares:")
for i, lang in enumerate(linguagens, start=1):
    print(f"  {i}. {lang}")

# List comprehension
quadrados = [x**2 for x in range(1, 11)]
print("\nQuadrados de 1 a 10:", quadrados)
```

## 4. Funções

```python
def calcular_media(valores):
    """Calcula a média aritmética de uma lista de valores."""
    if not valores:
        return 0
    return sum(valores) / len(valores)


def fatorial(n):
    """Calcula o fatorial de n usando recursão."""
    if n <= 1:
        return 1
    return n * fatorial(n - 1)


notas = [7.5, 8.0, 9.5, 6.0, 8.5]
print(f"Notas: {notas}")
print(f"Média: {calcular_media(notas):.2f}")

print(f"\nFatorial de 5: {fatorial(5)}")
print(f"Fatorial de 10: {fatorial(10)}")

# Lambda e função de ordem superior
dobrar = lambda x: x * 2
dobrados = list(map(dobrar, range(1, 6)))
print(f"\nDobrando [1..5]: {dobrados}")

pares = list(filter(lambda x: x % 2 == 0, range(1, 21)))
print(f"Números pares de 1 a 20: {pares}")
```

## 5. Programação Orientada a Objetos

```python
class Funcionario:
    """Representa um funcionário de uma empresa."""

    empresa = "Tech Corp"

    def __init__(self, nome, cargo, salario):
        self.nome = nome
        self.cargo = cargo
        self.salario = salario

    def reajuste(self, percentual):
        """Aplica um reajuste percentual ao salário."""
        self.salario *= (1 + percentual / 100)

    def __repr__(self):
        return f"Funcionario(nome={self.nome!r}, cargo={self.cargo!r}, salario=R${self.salario:.2f})"


class Desenvolvedor(Funcionario):
    """Especialização de Funcionario para desenvolvedores."""

    def __init__(self, nome, salario, linguagens):
        super().__init__(nome, "Desenvolvedor", salario)
        self.linguagens = linguagens

    def __repr__(self):
        base = super().__repr__()
        return f"{base}, linguagens={self.linguagens}"


dev = Desenvolvedor("Mateus", 8000, ["Python", "JavaScript", "SQL"])
print(dev)
dev.reajuste(10)
print(f"Após reajuste de 10%: {dev}")
```

## 6. Tratamento de Exceções

```python
def dividir(a, b):
    """Divide dois números com tratamento de erro."""
    try:
        resultado = a / b
    except ZeroDivisionError:
        print(f"Erro: divisão de {a} por zero não é permitida.")
        return None
    except TypeError as e:
        print(f"Erro de tipo: {e}")
        return None
    else:
        return resultado
    finally:
        print(f"Operação dividir({a}, {b}) concluída.")


print(dividir(10, 2))
print()
print(dividir(10, 0))
print()
print(dividir(10, "2"))
```

## 7. Trabalhando com Arquivos

```python
import json

# Escrevendo um arquivo JSON
dados = {
    "projeto": "Follow-up Python",
    "autor": "Mateus",
    "topicos": [
        "Variáveis",
        "Estruturas de Dados",
        "Funções",
        "POO",
        "Exceções"
    ],
    "versao": "1.0"
}

with open("/tmp/followup.json", "w", encoding="utf-8") as f:
    json.dump(dados, f, ensure_ascii=False, indent=2)

# Lendo o arquivo JSON
with open("/tmp/followup.json", "r", encoding="utf-8") as f:
    lido = json.load(f)

print(json.dumps(lido, ensure_ascii=False, indent=2))
```

## 8. Bibliotecas Populares

```python
from datetime import datetime, timedelta
import random
import statistics

# datetime
agora = datetime.now()
print(f"Data/hora atual: {agora.strftime('%d/%m/%Y %H:%M:%S')}")

daqui_30_dias = agora + timedelta(days=30)
print(f"Daqui a 30 dias: {daqui_30_dias.strftime('%d/%m/%Y')}")

# random e statistics
amostras = [random.gauss(50, 10) for _ in range(1000)]
print(f"\nAmostra aleatória (1000 valores, gauss μ=50, σ=10):")
print(f"  Média:    {statistics.mean(amostras):.2f}")
print(f"  Mediana:  {statistics.median(amostras):.2f}")
print(f"  Desvio:   {statistics.stdev(amostras):.2f}")
print(f"  Mínimo:   {min(amostras):.2f}")
print(f"  Máximo:   {max(amostras):.2f}")
```
